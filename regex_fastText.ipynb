```python
def preprocess(text):
  text = re.sub(r'[^\w\s]',' ',text)
  text = re.sub("[ \n]+"," ",text)
  return text.strip().lower()
```

```python
df1 = pd.read_csv("/content/ecommerce_dataset.csv",names=['category','description'],header=None)
print(df1.shape)
```

```python
df1.head()
```

```python
df1.category.value_counts()
```

```python
df1.dropna(inplace=True)
df1.shape
```

```python
df1.category.replace("Clothing & Accessories","Clothing_Accessories",inplace=True)
df1.category.unique()
```

```python
df1['category'] = '__label__' + df1['category'].astype(str)    # label as a prefix
df1.head()
```

```python
df1['category_description'] = df1['category'] + ' '+ df1['description']
df1.head()
```

```python
df1['category_description'][0]
```

```python
import re

def preprocess(text):
    text = re.sub(r'[^\w\s\']',' ',text)
    text = re.sub(r' +',' ',text)
    return text.strip().lower()   #strip remove leading spaces
```

```python
preprocess("  VIKI's | Bookcase/Bookshelf (3-Shelf/Shelve,white) | ? . hi")
```

```python
df1['category_description'] = df1['category_description'].map(preprocess)  # on entire category description col
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df1,test_size=0.2)
```

```python
train.shape,test.shape
```

```python
train.head()
```

```python
test.head()
```

```python
# will convert train and test to text file
train.to_csv("ecommerce.train",columns=["category_description"],index=False,header=False)
test.to_csv("ecommerce.test",columns=["category_description"],index=False,header=False)
```

```python
model = fasttext.train_supervised(input="ecommerce.train")
model.test("ecommerce.test")
# 1st one is precision, 2nd is recall
```

```python
model.predict("wintech assemble desktop pc cpu 500 gb sata hdd 4 gb ram intel c2d processor 3")
```

```python
model.predict("think and grow rich deluxe edition")
```
