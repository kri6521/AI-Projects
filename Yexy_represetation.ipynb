Varoius ways for converting text into vector - label encoding, one hot encoding, bag of words, tf-idf, word embeddings

They are not use in ml or nlp due to Disadvantages of
Label encoding -
one hot encoding - Doesn't capture meaning of the word in the accurate way, Waste lots of computer resources, suffers out of vocabulary prob, no fixed length representation

Bag of words = you have the vocabulary you get the word count for each of the words and create a vector which is also called count vectorizer.  

limitations of bag pf words: sparse representation which consumes a lot of memory in the comp, doesn't capture the meaning of the sentence accurately


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/mail_data.csv')
df.head()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)  #on category col it will apply some transformation and make a new col
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
X_train_cv = v.fit_transform(X_train.values)  #by values, it'll convert it to numpy array
X_train_cv
```

ML MODEL

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))   # truth nd the prediction
```

```python
emails = [
    'Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&Cs apply 08452810075over18'
]

emails_count = v.transform(emails)
model.predict(emails_count)
#spam
```

```python
# all the above steps can be done easily by creating pipeline
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```
